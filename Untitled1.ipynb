```python
import asyncio
from datetime import date

from infra.ingestion.adapters import EquityIngestionRequest, PolygonEquityAdapter
from infra.ingestion.adapters import PolygonRESTClient  # real client
from infra.ingestion.data_pipeline import EquityIngestionPipeline

rest_client = PolygonRESTClient(api_key="YOUR_POLYGON_KEY")
adapter = PolygonEquityAdapter(rest_client=rest_client)
pipeline = EquityIngestionPipeline(adapter=adapter)

request = EquityIngestionRequest(
  vendor="polygon",
  symbols=("AAPL", "MSFT"),
  start_date=date(2024, 1, 1),
  end_date=date(2024, 1, 5),
  flat_file_uris=(),
  options={},
)

loop = asyncio.get_running_loop()
manifest = await loop.run_in_executor(
  None,
  lambda: pipeline.run(request, run_id="demo_equity_ingest"),
)
manifest
```
